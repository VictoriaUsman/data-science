```python
#load importantant libraries
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# View first rows
diamonds.head()
```

Univariate Analysis (One Variable)

```python
# Univariate Analysis for price
plt.hist(diamonds['price'])
plt.title('Distribution of Diamond Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

# I chose the variable 'price' for Univariate Analysis.
# This histogram shows the distribution of diamond prices.

Multivariate Analysis (Two Variables)

```python
# Multivariate Analysis for carat and price
plt.scatter(diamonds['carat'], diamonds['price'])
plt.title('Carat vs Price')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()
```

# I chose 'carat' and 'price' for Multivariate Analysis.
# This scatter plot shows the relationship between diamond size and price.
