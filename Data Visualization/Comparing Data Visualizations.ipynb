```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv('countries.csv')
```

```python
# Filter for the year 1952
data_1952 = df[df.year == 1952]
```

```python
# Create subsets for Asia and Africa
asia_1952 = data_1952[data_1952.continent == 'Asia']
africa_1952 = data_1952[data_1952.continent == 'Africa']
```

```python
# Create subplots to compare the distributions
# We use 2 rows and 1 column for the layout
plt.figure(figsize=(10, 8))
```

```python
# Subplot 1: Asia
plt.subplot(2, 1, 1)
plt.hist(asia_1952.lifeExp, bins=20, color='skyblue', edgecolor='black')
plt.title('Life Expectancy Distribution (1952)')
plt.ylabel('Asia (Freq)')
```

```python
# Subplot 2: Africa
plt.subplot(2, 1, 2)
plt.hist(africa_1952.lifeExp, bins=20, color='orange', edgecolor='black')
plt.ylabel('Africa (Freq)')
plt.xlabel('Life Expectancy')
```

```python
# Purpose: This script loads global development data and uses histograms 
# to visually compare the health outcomes (life expectancy) of two 
# different continents in the same historical year.
```
